```python
import pandas as pd
import numpy as np
import glob
```

```python
csv_file_list = ["states0.csv", "states1.csv", "states2.csv", "states3.csv", "states4.csv", "states5.csv", "states6.csv", "states7.csv", "states8.csv", "states9.csv"]

# get data file names
#path =r'C:\DRO\DCL_rawdata_files'
filenames=glob.glob('states?.csv')
#filenames = glob.glob(path + "/*.csv")
list_of_dataframes = []
for filename in filenames:
    list_of_dataframes.append(pd.read_csv(filename))

us_census = pd.concat(list_of_dataframes,ignore_index=True)
```
